# Bankruptcy Prediction

The dataset consists of financial information for ~6400 companies from the period 1979 to 2004. In this notebook I will explore the dataset and compare several classification models to predict bankruptcy.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import cross_val_score, train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
from sklearn.metrics import precision_recall_fscore_support, roc_curve, accuracy_score, roc_auc_score

import warnings
warnings.filterwarnings("ignore")
```


Features:
- `EPS` – Earnings Per Share
- `Liquidity` – Working Capital/Total Assets
- `Profitability` – Retained Earnings/Total Assets
- `Productivity` – EBIT/Total Assets
- `Leverage Ratio` – (Total Long-term debt + Debt in Current liabilities)/Stockholders Equity
- `Asset Turnover` – Sales/ Total Assets
- `Operational Margin` – EBIT/Sales
- `Market Book Ratio` – (Price Close Annual Fiscal * Common Shares Outstanding)/Book Value Per Share
- `Asset Growth` – Change in assets from previous year
- `Sales Growth` – Change in sales from previous year
- `Employee Growth` – Change in employees from previous year
- `Tobin’s Q` – (Total market value of company + liabilities)/ (Total asset or book value + liabilities)

Target:
- `BK` – Company bankrupt or not

## 2. Data Processing

### Handle Missing Data
Given the large number of NaNs (~12.6%), we need to need to come up with a strategy to impute the missing values. The strategy is as follows:
1. Look at the data and figure out which rows correspond to a company, as we are not explicitly given company details. The idea is that everytime the `year` goes back in time (eg. goes from 2014 to 1996) we are looking at a new company
2. For each missing value, look within the company's historical data and fill in the missing year with the nearest year. This is dones using forward fill and backward fill
3. For any remaining missing values - i.e. the company doesn't have any historicaly record of a feature - we impute using the mean from the dataset

```python
#Step 1: Add Company_ID based on year
data['Company'] = data['Year'].diff() # Idea is that new company every time the change in years is Negative

#Identify unique companies
company_id = 100
for idx, row in data.iterrows():
    if data.loc[idx,'Company'] < 0 : #I.e. This is a new company
        company_id += 1
    data.loc[idx,'Company'] = company_id

#Step 2: Impute missing values with closest year match

data['Company2'] = data['Company'] #Duplicating so that the Comapny column carries over
data['Company3'] = data['Company'] #Duplicating so that the Comapny column carries over

data = data.groupby(by='Company2').transform (lambda x: x.fillna(method='ffill')) # method='ffill'
data = data.groupby(by='Company3').transform (lambda x: x.fillna(method='bfill')) # method='ffill'

data.head()
```

There are still some entries that have missing values. Most of these are in `Operational_Margin`. 

```python
print ("Total number of unique companies: ", len(data.Company.unique()))
print ("Number of unique companies with missing data:", len(data[data.isnull().any(axis=1)].Company.unique()),'\n')

data[data.isnull().any(axis = 1)].info()
```

Note that ~6% of these missing data contain bankrupcy. This is 6 times the 1% bankrupcy rate in entire dataset. Therefore, I made the choice to fill these missing values with the mean

```python
#Step 3: Fill in the resto of mssing values with the mean
data = data.apply(lambda x: x.fillna(x.mean()),axis=0)
data.info()
```

### Split data into train and test sets

Before we begin, I will split the datase into two chunks:
1. 70% for a *train* set - this is used for hyperparameter turning
2. 30% for a a *Test* set - that is  used for testing and validation

```python
#Splitting the data df into X and y lists
X_prior = data[feature_cols]
y_prior = data.BK
print ("shape of X_prior: ", X_prior.shape)
print ("shape of y_prior: ", y_prior.shape)
```

```python
#Split the dataset 30% for holdout set
X_train, X_test, y_train, y_test = train_test_split(X_prior, y_prior, test_size = 0.3, random_state = 42, stratify = y_prior )
print ("shape of X_train: ", X_train.shape)
print ("shape of y_train: ", y_train.shape)
print ("shape of X_test: ", X_test.shape)
print ("shape of y_test: ", y_test.shape)
```

```python
#Check if stratify worked. The means should rougly be the same
print(y_train.mean())
print(y_test.mean())
```

### Standardize
Next step is to normalize/standardize the data

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Upsampling using SMOTE
Given the high class imbalance (1% positive class), I decided to upsample using SMOTE prior to building the model. Downsampling would have resulted in thowing away a lot of data.

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE (random_state = 42)
print ('ORIGINAL DATA ')
print ("shape of X_train: ", X_train.shape)
print ("shape of y_train: ", y_train.shape)
print ("Number of positive class:", y_train.sum())
print ("Number of negative class:", len(y_train)-y_train.sum())

print ('\nUPSAMPLED DATA ')
X_train_res, y_train_res = sm.fit_sample(X_train,y_train)
print ("shape of X_train_res: ", X_train_res.shape)
print ("shape of y_train_res: ", y_train_res.shape)
print ("Number of positive class:", y_train_res.sum())
print ("Number of negative class:", len(y_train_res)-y_train_res.sum())
```

```python
#Export to csv
temp = pd.DataFrame(X_train_res, columns = feature_cols).join(pd.Series(y_train_res, name = "BK"))
temp.to_csv('./data/data_Standardized_SMOTE.csv')
#Export holdout set to csv
temp = pd.DataFrame(X_test, columns = feature_cols).join(y_test.reset_index(drop=True))
temp.to_csv('./data/test_data.csv')
del(temp)
```

## 3. Machine Learning Models

I chose to compare the performance for 3 different classifiers:
1. Support Vector Machines:
    - with RBF Kernel
    - with Linear Kernel
2. Logistic Regression
3. Random Forests


```python
#Optional to import data
train = pd.read_csv('./data/data_Standardized_SMOTE.csv')
test = pd.read_csv('./data/test_data.csv')

X_train_res = train.loc[:,"Tobins_Q":"Employee_Growth"]
y_train_res = train.loc[:,"BK"]

X_test = test.loc[:,"Tobins_Q":"Employee_Growth"]
y_test= test.loc[:,"BK"]
```

### 3.1 GridSearch 

I'll use GridSearch for hyperparameter tuning. 
Note: Since SVM takes a long time to grid-search I will train without probabilistic output. Later, I will re-train the the best model using `probability = True` to generate the ROC curve

#### SVM with RBF Kernel


```python
clf1 = svm.SVC(kernel = 'rbf')
param_svm_rbf ={'C': [1, 10, 100],
            'gamma': [0.001, 0.0001]
           }
grid_svm_rbf = GridSearchCV(clf1, param_svm_rbf, cv = 5, scoring='roc_auc', n_jobs = -1,  verbose = 2)

grid_svm_rbf.fit(X_train_res, y_train_res)
```

```python
grid_svm_rbf.best_params_
```

#### SVM with Linear Kernel

```python
clf2 = svm.SVC(kernel = 'linear')
param_svm_lin ={'C': [1, 10, 100],
             'class_weight':[None, 'balanced']
            }
grid_svm_lin = GridSearchCV(clf2, param_svm_lin, cv = 5, scoring='roc_auc', n_jobs=-1, verbose = 2)

grid_svm_lin.fit(X_train_res, y_train_res)
```

```python
grid_svm_lin.best_params_
```

#### Random Forest

```python
clf3 = RandomForestClassifier(criterion='entropy')

param_forest = {
   'bootstrap': [True,False],
   'max_depth': [10, 25, 50],
   'max_features': ['log2', 'sqrt'],
#   'min_samples_leaf': [1, 2, 4],
#   'min_samples_split': [2, 5, 10],
   'n_estimators': [10, 50, 100, 200]
}

grid_forest = GridSearchCV(clf3, param_forest, scoring = 'roc_auc', n_jobs = -1, cv = 5, verbose = 2)

grid_forest.fit(X_train_res, y_train_res)
```

```python
grid_forest.best_params_
```

#### Logistic Regression

```python
clf4 = LogisticRegression()
param_logreg = {
         #liblinear uses l1 penalty, lbfgs uses l2 penalty
        'penalty': ['l1', 'l2'],
        'C': np.logspace(-5,8,10),
        'class_weight': [None, 'balanced']
    }

grid_logreg = GridSearchCV(clf4, param_logreg, scoring='roc_auc', n_jobs = -1, cv=5, verbose = 2)

grid_logreg.fit(X_train_res, y_train_res)
```

```python
grid_logreg.best_params_
```

### 3.2 Compare Models
#### GridSearch Scores
Let's look at the average AUC score form GridSearch for each classifier.

```python
print("Average Test Scores:")
print("SVM RBF: ", np.mean(grid_svm_rbf.cv_results_['mean_test_score']))
print("SVM Linear: ", np.mean(grid_svm_lin.cv_results_['mean_test_score']))
print("Random Forest: ", np.mean(grid_forest.cv_results_['mean_test_score']))
print("Logistic Regression: ", np.mean(grid_logreg.cv_results_['mean_test_score']))
```

`Random Forest` has **0.995** AUC score on the train data. The next best classifier is `SVM Linear` with AUC of **0.71**. `SVM RBF` and `Logistic Regression1` have similar AUC scores ranging **~0.66 to 0.67**.

Is Random Forest the best classifier for this task? Or could it simply be overfitting and memorizing the Train data. To have a better comparision we need to evaluate the classifiers on the TEST data.


#### ROC Curve 
I will use the ROC curve with AUC to compare the performance of each classifier on the **Test** data set. 

Since I didn't initiate probabilistic output, I will re-run SVM with the best paramemters by passing in
`probability = True`


```python
svm_rbf = svm.SVC(kernel='rbf', C = 100, gamma = 0.001, probability = True)
svm_rbf.fit(X_train_res, y_train_res)
svm_lin = svm.SVC(kernel='linear', C=100, probability = True)
svm_lin.fit(X_train_res, y_train_res)
```

```python
# Generate probabilistic predictions for each classifer
y_pred_proba_svm_rbf = svm_rbf.predict_proba(X_test)[:,1]
y_pred_proba_svm_lin = svm_lin.predict_proba(X_test)[:,1]
y_pred_proba_forest = grid_forest.predict_proba(X_test)[:,1]
y_pred_proba_logreg = grid_logreg.predict_proba(X_test)[:,1]
```

```python
# Generate ROC curve values: fpr, tpr, thresholds
models = {'SVM-RBF': y_pred_proba_svm_rbf, 
          'SVM-Linear':y_pred_proba_svm_lin,
          'Random Forest':y_pred_proba_forest, 
          'Logistic Regression': y_pred_proba_logreg}

plt.figure(figsize=(10,8))
plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('1-Specificity(False Positive Rate)')
plt.ylabel('Sensitivity(True Positive Rate)')
plt.title('ROC Curve')
for name, y_pred_proba in models.items():
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    auc = roc_auc_score(y_test,y_pred_proba)
    # Plot ROC curve
    plt.plot(fpr, tpr, label='%s (area = %0.2f)' % (name, auc))

plt.legend(loc="lower right")

plt.show() 
```

`Random Forest` is still the best classifier accoring to its ROC value of **0.92**! However, notice that the AUC score dropped from 0.995 to 0.92. The other three classifiers have very similar performance with AUC score of ~**0.62**. Interestingly, `Logistic Regression` which had a high train AUC of **0.71** didn't generalize well on the Test data (test AUC **0.62**).

#### Classification Report

Finally I will print out a classification report for each classifier. For this, I am chosing a treshold value from the ROC curve above which corresponds to the highest recall for a `False Positive Rate <= 0.2`.


```python
models = {'SVM_RBF': y_pred_proba_svm_rbf, 
          'SVM_Linear':y_pred_proba_svm_lin,
          'Random_Forest':y_pred_proba_forest, 
          'Logistic_Regression': y_pred_proba_logreg}

roc_scores = pd.DataFrame(columns = ['model','fpr','tpr','thresholds'])

#Create a dataframe that stores the tpr,fpr,thresholds for each classifier
for name, y_pred_proba in models.items():
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    df = pd.DataFrame({'tpr':tpr,'fpr':fpr,'thresholds':thresholds})
    df['model'] = name
    roc_scores = pd.concat([roc_scores,df], ignore_index = True, sort=True)
    del(df)

#optional - for readability 
roc_scores.sort_values(by=['model','tpr'], ascending = False, inplace = True)

# Identify Threshold values where False Positive Rate <= 0.2
threshold_svm_rbf = roc_scores.loc[(roc_scores.model=='SVM_RBF')&(roc_scores.fpr<=0.2),'thresholds'].head(1).values[0]
threshold_svm_lin = roc_scores.loc[(roc_scores.model=='SVM_Linear')&(roc_scores.fpr<=0.2),'thresholds'].head(1).values[0]
threshold_forest = roc_scores.loc[(roc_scores.model=='Random_Forest')&(roc_scores.fpr<=0.2),'thresholds'].head(1).values[0]
threshold_logreg = roc_scores.loc[(roc_scores.model=='Logistic_Regression')&(roc_scores.fpr<=0.2),'thresholds'].head(1).values[0]
```

Calculate Precision, Recall, F1 Score, Accuracy, and AUC scores for each model using the thresholds calculated above

```python
models = {'SVM_RBF': [y_pred_proba_svm_rbf, threshold_svm_rbf], 
          'SVM_Linear': [y_pred_proba_svm_lin, threshold_svm_lin ],
          'Random_Forest': [y_pred_proba_forest,threshold_forest ], 
          'Logistic_Regression': [y_pred_proba_logreg, threshold_logreg]}

summary = pd.DataFrame(columns = ['Model','Precision','Recall','Fscore', 'Accuracy', 'AUC'])

for name, model in models.items():
    precision, recall, fscore, _ = precision_recall_fscore_support(y_test, model[0] >= model[1])
    accuracy = accuracy_score (y_test, model[0]  >= model[1])
    auc = roc_auc_score (y_test ,model[0] )
    
    summary = summary.append({'Model':name,'Precision':precision[1],'Recall':recall[1],'Fscore':fscore[1], 'Accuracy':accuracy, 'AUC':auc}, ignore_index= True)
    
summary.set_index('Model', inplace=True)
summary
```

#### Conclusion
Given that we are interested in identifying bankruptcies, we are interested in model that has a high Recall score. In other words, we want our model to identify as many bankruptices as possible. Of the four models, `Random Forest` had the highest recall score of **0.87**. Moreover, `Random Forest` had a very high AUC score of **0.92**, meaning that it strikes a good balance between recall and the False positive rate, making it the best performing model on the list.
